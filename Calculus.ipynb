```python
x_points = np.linspace(-10,10,50)
```

```python
def fun_1(x):
    y_list = []
    for elem in x:
        y = math.pow(elem,2) + (elem * 3) - 1
        y_list.append(y)
    return y_list
```

```python
y_points = fun_1(x_points)
```

```python
plt.plot(x_points,y_points)
plt.grid()
plt.show()
```

```python
def onclick(event):
    ix, iy = event.xdata, event.ydata
    print(f'x = {ix}, y = {iy}')
```

```python
fig = plt.figure()
cid = fig.canvas.mpl_connect('button_press_event', onclick)
plt.plot(x_points,y_points)
plt.show()
```
