```python
msg_len =len('Message transfer to raw bkp topic')
files = [ log_dir+"/"+f for f in os.listdir(log_dir)]
for f in files:
    with open(f,'r', encoding='UTF-8') as reader:
        for line in reader:
            if 'Message transfer to raw bkp topic' in line:
                line = line[line.index('Message transfer to raw bkp topic')+msg_len:]
                csv_line = ",".join(line.split("#"))
                with open(KAFKA_MSG,'a', encoding='UTF-8') as writer:
                    writer.write(csv_line)
```

```python
list_hash =[]
counter=0
with open(missing_trip_vin_kafka_data,'r',encoding='UTF-8') as reader:
    for line in reader:
        hash_1 = hash(line)
        if hash_1 not in list_hash:
            list_hash.append(hash_1)
        counter= counter+1
        
print("counter>> ",counter)
print("list len >> ",len(list_hash))
```
