```python
from folium.plugins import Draw

# Step 3: Create a map object
m = folium.Map(location=[51.4347154,5.5041018], zoom_start=12)
# Step 4: Create a Draw object
draw = Draw(export=True)
draw.add_to(m)

# Step 5: Define polygon coordinates
polygon_coords = [
(51.4347154,5.5041018),
(51.4315463,5.5028398),
(51.4251616,5.5428944),
(51.4318233,5.5343023),
(51.4347035,5.5042004)
]

# Step 6: Create a polygon object and add it to the map
polygon = folium.Polygon(locations=polygon_coords, color='blue', fill=True, fill_color='blue')
polygon.add_to(m)
```

```python
data_path = "C:/Users/a818517/Documents/bugs/32054/enterZone_4.txt"
plot_data = None
plot_data = []
with open(data_path,'r') as reader:
    for line in reader:
        try:
            data = line.strip()
            json_dict = json.loads(data)
            if json_dict['GPSLatitude'] != "255.0":
                plot_data.append((json_dict['EvtDateTime'],json_dict['GPSLatitude'],json_dict['GPSLongitude']))
        except:
            pass
```

```python

line_coords = [(float(d[1]),float(d[2])) for d in plot_data ]
polyline = folium.PolyLine(locations=line_coords, color='red', weight=2)
polyline.add_to(m)

## Adding marker
for idx,each in enumerate(line_coords):
    folium.Marker(each,tooltip=f"{idx} > {plot_data[idx]}").add_to(m)
m
```
